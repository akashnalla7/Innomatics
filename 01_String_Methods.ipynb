3. What is the difference between string.find() and string.index() ? Explain with examples.


- find() and index() return the index position of the first occurrence of the substring

- Key difference : find() produces -1 as output if it is unable to find the substring, where index() throws a ValueError.

```python
# Examples

string1 = "This book is really great one."

# find()

print("---- FIND ----")
print(string1.find("is")) # It returns the index of the substring.
print(string1.find("an")) # Here it unable to find the substring then it returns -1.

# index()
print("---- INDEX ----")
print(string1.index("is")) # It returns the index of the substring.
print(string1.index("an")) # Here it unable to find the substring then throws a ValueError
```

5.  Given two strings, first_name and last_name, write a function that concatenates them to form a full name with a space in between. Ensure the resulting full name has the first letter of each word capitalized correctly.


```python
def fullname(first_name, last_name):

  full_name = first_name + ' ' + last_name
  full_name = full_name.title()

  return full_name

fullname("akash","nalla")
```

6. Check if the string s (s = "traveling") is a palindrome and assign the result (True/False) to the variable is_palindrome.

```python
s = "traveling"

is_palindrome = s is s[::-1]

print(is_palindrome)
```
